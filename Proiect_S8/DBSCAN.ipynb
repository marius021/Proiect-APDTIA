```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("CC GENERAL.csv")
```

```python
print(dataset.shape)
```

```python
dataset.head()
```

```python
len(dataset['CUST_ID'].unique())
```

Stergerea coloanei **CUST_ID**

```python
dataset.drop('CUST_ID', axis = 1, inplace = True)
```

```python
dataset.info()
```

Inlocuirea valorilor lipsa cu valorile medii

```python
dataset = dataset.fillna(dataset.mean())
```

```python
dataset.info()
```

```python
dataset.describe()
```

Normalizarea datelor

```python
from sklearn.preprocessing import StandardScaler, normalize

scaler = StandardScaler()
dataset = pd.DataFrame(normalize(scaler.fit_transform(dataset)))
```

Reducere dimensionalitatii - Principal Component Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
dataset = pd.DataFrame(pca.fit_transform(dataset))
dataset.columns = ['Column 1', 'Column 2']
```

Aplicarea DBSCAN

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 0.05, min_samples=10)
model.fit(dataset)
```

Labels - categorie pentru fiecare cluster, iar "outliers" vor fi clasificate ca -1




```python
labels = model.labels_
np.unique(labels)
```

Afisare cate puncte (obiecte) sunt in cluser si cate anomalii

```python
points = list(labels).count(0)
print("Puncte in cluster:", points)
```

```python
outliers = list(labels).count(-1)
print("Anomalii:", outliers)
```

Afisare grafica a clusterului si anomaliilor

```python
colors = {}
colors[-1] = 'r'
colors[0] = 'b'
b = plt.scatter(dataset['Column 1'], dataset['Column 2'], color = 'b')
r = plt.scatter(dataset['Column 1'], dataset['Column 2'], color = 'r')
cvector = [colors[label] for label in labels]

plt.figure(figsize = (9, 9))
plt.scatter(dataset['Column 1'], dataset['Column 2'], c = cvector )
plt.legend((b, r), ('Cluster', 'Anomalii'))
plt.show()
```
