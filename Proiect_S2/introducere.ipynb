```python
import numpy as np
```

```python
speed = [99, 12, 88, 72, 43, 54, 66, 99]
```

```python
medie = np.mean(speed)
# Valoare medie = media aritmetica a tuturor elementelor
print(medie)
```

```python
# Val mediana - valoarea din mijloc dupa sortarea crescatoare a valorilor
mediana = np.median(speed)
print(mediana)
```

```python
#Mod - cea mai frecvventa valoare

from scipy import stats
mode = stats.mode(speed)

print(mode)
```

```python
"""
Varianta - indica cat de raspandite sunt valorile

1. Se determina val medie
2. Pentru fiecare val din set se scade val medie
    99 - 66.625 = 32.375
    12 - 66.625 = -53....
    ...
3. Se ridica la ^2 toate val determinate la pct 2.
    32.375^2 = ...
4. Var se det prin calc a mediei aritmetice de la pct 2

"""

varianta = np.var(speed)
print(varianta)
```

```python
"""
Abaterea standard (Standard deviation) - indica raspandirea valorilor 
- este corelata cu varianta ( radicalul variantei)

"""

std = np.std(speed)
print(std)
```

```python
#Percentile
set = [10, 90, 86, 34, 91, 13, 15, 78, 9, 44]
p50 = np.percentile(set, 50)
print(p50)
```
