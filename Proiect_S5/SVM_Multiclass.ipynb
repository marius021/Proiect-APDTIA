```python
from sklearn import datasets
```

```python
dataset = datasets.load_iris(as_frame = True)
```

```python
print("Caracteristici:", dataset.feature_names)
```

```python
print("Target:", dataset.target_names)
```

Definirea vectorilor X (caracteristici) si y (tinta)

```python
X = dataset.data
y = dataset.target
```

Analiza setului de date

```python
X.describe()
```

```python
X.shape
```

```python
y.shape
```

```python
X.head()
```

```python
#Nr de elemente avand clasa 0
(y == 0).sum()
```

```python
#Nr de elemente avand clasa 1
(y == 1).sum()
```

```python
#Nr de elemente avand clasa 2
(y == 2).sum()
```

Impartirea setului de date in train/test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

Crearea si antrenarea clasificatorului cu kernel liniar

```python
from sklearn import svm
model = svm.SVC(kernel="linear")
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

Evaluarea modelului

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(metrics.classification_report(y_test, y_pred))
```

Testare cu kernel RBF

```python
model = svm.SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(metrics.classification_report(y_test, y_pred))
```

Testare cu kernel Polinomial

```python
model = svm.SVC(kernel = "poly")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(metrics.classification_report(y_test, y_pred))
```
